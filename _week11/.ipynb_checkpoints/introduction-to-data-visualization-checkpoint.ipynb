## Ex 1: Using line plots to visualize the rise and fall of the "triple-decker" (3-volume) Victorian novel between 1837-1901 
  
Hypothesis: Literary critics love to talk about the rise of the "triple-decker" or 3-volume novel ðŸ“š in nineteenth-century British lending libraries. The format, so the story goes, became popular because circulating libraries were some of the chief buyers of novels, and having a novel in 3 parts increases threefold the number of people you might lend that same novel out to (thus cutting down on the number of copies the library had to hold)

>**Our question: When did triple-decker novels overtake other formats for novel publishing?  
And how did they stack up in comparison to the one-, two- and four (or more)-volume novels?**

[At the Circulating Library, 1837-1901](https://github.com/sceckert/Data-and-Literary-Study-Spring2022/tree/main/_datasets/At-the-Circulating-Library-Dataset/) is a dataset of novels published in Great Britain that comes from the [At the Circulating Library Project](https://www.victorianresearch.org/atcl/index.php), run by Troy Basset. For more on the various bibliographies and primary sources used to determine the full list of one-, two- and three+ volume noves, see the project's [Documentation page](https://www.victorianresearch.org/atcl/documentation.php). The dataset has a few different sets of data.  We're going to focus on the "titles.csv" dataset, which contains information about the title, format (how many volumes), publishing locaiton, publisher, and date of publication, along with links to the author information.


```python
circulating_library_df = pd.read_csv('../_datasets/At-the-Circulating-Library-Dataset/titles.csv')
```

```python
circulating_library_df
```

#### Using `groupby()` to group by volumes and publication year

```python
circulating_library_df_grouped_by_vol = circulating_library_df.groupby(['volumes', 'publication_year'])['title_id'].count().reset_index()
```

#### Using `rename()` to rename a column 

```python
# Rename the column "title_id" to more accurately reflect what it now contains: the count of the number of records
circulating_library_df_grouped_by_vol.rename(columns = {'title_id':'record_count'}, inplace = True)
circulating_library_df_grouped_by_vol
```

#### Using `.pivot()` to reshape our dataframe
Here, we're pivoting so that the values (eg, 1, 2, 3, or 4) that had been rows in the "volumes" column are now individual columns. This will make it easier for us to plot line graph representations of our data over time.

We're also going to set the "publication_year" as the index (this will make it easier so that our plot function knows to assign the publication_years to the x axis).

```python
pivoted_circulating_library_df_grouped_by_vol = circulating_library_df_grouped_by_vol.pivot(index='publication_year', columns='volumes', values='record_count').fillna(0)
pivoted_circulating_library_df_grouped_by_vol.head()
```

#### Plotting our reshaped dataframe as a line graph
See below: we're going to use the `.plot.line()` function in its more general form as `.plot(kind='line')`.

Notice how because we've assigned our dates to the index, we don't need to specify the x and y axis. THis is because the `plot.line()` or `.plot(kind='line')` function assumes that the index is the x axis. According the [pandas documentation on plot.line](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.line.html), here are some aditional parameters we could specify:
> "x=" Allows plotting of one column versus another. If not specified, the index of the DataFrame is used.
ylabel or position, optional
> "y="Allows plotting of one column versus another. If not specified, all numerical columns are used.


```python
pivoted_circulating_library_df_grouped_by_vol.plot(kind='line',\
                             # Typing \ backslash, followed by ENTER makes our code to make it easier to read
                              # it tells Python that the first line of code continues onto the next line.
                                                   legend= True,\
                                                   title="How Many Victorian Novels Were Published (and in What Format) between 1837-1901?" )
```

What if we wanted to know the rough breakdown of genres in the "little" journals that constituted an important set of materials for English-language coterie modernism during the 1890s-1920s? The Modernist Journals Project (MJP) has 

For the purpose of this example, we're going to look at data from just two journals: [The Egoist](https://modjourn.org/journal/egoist/) all six volumes (and 74 issues) from 1914 to 1919, and [Poetry](https://modjourn.org/journal/poetry/): just the first 20.5 volumes (and 123 issues) from 1912 through 1922.

I created this dataset by concatenating the [*Poetry* magazine TSV file of every title](https://sourceforge.net/projects/mjplab/files/Poetry%20Magazine/) and the [*The Egoist* TSV file of every title](https://sourceforge.net/projects/mjplab/files/Egoist/) and cleaning up tab spacing errors.
```python
modernist_journals_df = pd.read_csv('../_datasets/Modernist-Journals-Project/Poetry_and-Egoist-everytitle.txt', delimiter='\t', encoding='utf-8')
```

```python
modernist_journals_df.head()
```

### Using `.value_counts()` to get the number of unique rows with a given value
We'll use this to get a count of how many items of poetry, articles, letters, etc. there are in our *Poetry* and *The Egoist* dataset

```python
modernist_journals_df['genre'].value_counts()
```

###  Using `.assign()` and `.explode()` to split cells with multiple values on a delimiter 
Occasionally we may want to split up a cell that contains multiple values into distinct rows (like "," or ";" or " | "). The `assign()` method helps us do that, while the `explode()` method copies over the data to the newly created row. 

```python
# Notice the semicolon in "letters; poetry"? We have one work that was tagged as two genres
# Let's split that cell up, so we can add that work to both the "letters" and "poetry" counts
modernist_journals_df = modernist_journals_df.assign(genre=modernist_journals_df['genre'].str.split('; ')).explode('genre')
```

```python
mjp_genre_counts = modernist_journals_df['genre'].value_counts()
mjp_genre_counts
```

### Using a `.plot()` horizontal bar chart to compare the record counts in different categories

The general .`plot(kind='INSERT KIND HERE')` function can take objects that are not dataframes. In this case, we're going to pass the list of integers that we assigned to "mjp_genre_counts".

```python
mjp_genre_counts.plot(kind='barh', figsize=(10,10),\
                      title="Genres in POETRY and THE EGOIST, 1913-1919")
```

### Using a `.plot()` pie chart to compare the record counts in different categories

```python
mjp_genre_counts.plot(kind='pie', figsize=(10,10), \
                      title ="Genres in POETRY and THE EGOIST, 1913-1919")
```

# Data Visualization with Altair

Sometimes, the kinds of relationships between aspects of your dataset is more complex than a simple `pandas` plot can handle! 

We're going to learn how to create a simple interactive visualization using [Altair](https://altair-viz.github.io/), which is a Python library built on top of Vega and Vega-Lite - two visualization libraries for JavaScript. 

[`Altair`](https://altair-viz.github.io/) can take a lot more customization than the standard plotting library in `pandas` and can handle much more intricate visualizations. Another library that's also good for more complicated visualizations is [seaborn](https://seaborn.pydata.org/), which we used to generate the heatmaps for our topic model heatmap last week.


## Ex 3: Making a staked bar chart to visualize the breakdown in genres within each issue of *Poetry* and *The Egoist*

Let's say we want to visualize the breakdown of genres in each issue, by year and jouranl title.

First, we have to isolate the data that we want to plotâ€“â€“we don't want all of our dataframe. 

To do that, we're going to create a new dataframe ("modernist_journals_df_sorted) with just the data we want to plot use the `.groupby()` function to group by genre, journal title, and date, and selecting the "title" column. The next part of this line of code counts the number of records using `count()` on the title column (remember, these serve as proxy for the individual worksâ€“â€“if we want to know how many individual pieces were labled a given genre). Finally, the last piece of the code resets the index, so that the categories we grouped by are now columns again, rather than in the index.

```python
modernist_journals_df_sorted = modernist_journals_df.groupby(['genre', 'journal title', 'date'])['title'].count().reset_index()
modernist_journals_df_sorted 
```

```python
# Rename the column "title" to more accurately reflect what it now contains: the count of the number of records
modernist_journals_df_sorted.rename(columns = {'title':'record_count'}, inplace = True)
modernist_journals_df_sorted 
```

## Import Altair

```python
import altair as alt
```

### Plot our visualization
We're going to use Altair to plot, then I'll explain below what this peice of code is doing:

```python
chart = alt.Chart(modernist_journals_df_sorted,\
                  title='Works Published in POETRY and THE EGOIST by Genre, 1913-1919').mark_bar(size=3).\
encode(
   column='journal title',     
    x=alt.X('yearmonthdate(date):T', axis=alt.Axis(format='', title='Publication Date')
    ),
    y=alt.Y('record_count:Q', axis=alt.Axis(title='Number of works')),
    color='genre:N',
)

chart.display()
```

What did we just do? 

- We specify a new `Chart` class 
- We then specify the type of mark we are using (in this case, a `bar`. In Altair, we can use all types of marks to represent our data https://altair-viz.github.io/user_guide/marks.html. 
- We call `encoding` to specify what variable we want to represent on the x and y axis,  as well as specifying  how we want the data to be encoded (i.e., making sure that our "date" column is parsed as yearsmonthdate and as temporal data "T"), and adding labels to our x and y axis. We also use this to specify that we want the grid separated into two columns (here, based on the values in the "journal title" column), and we sepcify that we want our bar charts to be colored by the values in our "genre" column.
- Finally, we call `.display` to tell altair to display our visualization.

Altair has many fields for encoding https://altair-viz.github.io/user_guide/encoding.html

```python
# specify a new Chart class and set it equal to the variable "chart" (so that we can reference it more easily)
chart = alt.Chart(modernist_journals_df_sorted,\
                  #specify the title and the type of mark that we're using
                  title='Works Published in POETRY and THE EGOIST by Genre, 1913-1919').mark_bar(size=3).\
encode(
    # don't be confused by the word "column"! This next line separates our chart into TWO grids, one for each magazine
    column='journal title',     
    # specifies the x axis,  the format that we want to parse our "date" column, and the label for the axis
    x=alt.X('yearmonthdate(date):T', axis=alt.Axis(format='', title='Publication Date')
    ),
    # specifies the y axis and the label for the axis
    y=alt.Y('record_count:Q', axis=alt.Axis(title='Number of works')),
    # specfies that we want the segments of our bar chart colored according to the "genre" column
    color='genre:N',
)

chart.display()
```

## Ex 4: Let's make an interactive visualization of our genre data in Poetry and Egoist

We're going to follow the same steps as above: specifying a chart class, the type of mark, the encoding, and the properties.

But we're also going to add two new steps:

- We're going to specify a `tooltip`, which will create a tooltip to disply additional information about our datapoints when we hover over it.
- We're going to call `interactive` to specify that we want our visualization to be interactive.

Altair has built interactivity fields.

Altair ALSO has a function to save your chart (see below)

```python
chart = alt.Chart(modernist_journals_df_sorted,\
                  title='Works Published in POETRY and THE EGOIST by Genre, 1913-1919').mark_bar(size=3).\
encode(
   column='journal title',     
    x=alt.X('yearmonthdate(date):T', axis=alt.Axis(format='', title='Publication Date')
    ),
    y=alt.Y('record_count:Q', axis=alt.Axis(title='Number of works')),
    color='genre:N',
    tooltip=['genre', 'journal title', 'yearmonthdate(date):T', alt.Tooltip('record_count:N')]
).interactive()

chart.display()
```

## Ex 5: Saving an Altair chart as an HTML file
If you right click at the top of one of the above charts, you'll be able to download the chart as a static image. But what if we wanted to save an *interactive* version? To save a chart made in altair as an HTML file, simply use the `.save('your-name-for-your-chart.html')` operation. We're also going to tell altair to make our chart a little bigger by calling the `properties` specification. While we could add `.save('YourChartName.html')` at the end of the very end of our line of code, after the `properties()` specification.

But we because we set the entire Chart class equal to the variable "chart", we can also run the line "chart.save('MJP-works-in-poetry-and-the-egoist-by-genre.html')"



Click and run the cell below. Look in the lefthand panel of JupyterLabâ€“â€“ did you see your HTML file?

```python
chart = alt.Chart(modernist_journals_df_sorted,\
                  title='Works Published in POETRY and THE EGOIST by Genre, 1913-1919').mark_bar(size=3).\
encode(
   column='journal title',     
    x=alt.X('yearmonthdate(date):T', axis=alt.Axis(format='', title='Publication Date')
    ),
    y=alt.Y('record_count:Q', axis=alt.Axis(title='Number of works')),
    color='genre:N',
    tooltip=['genre', 'journal title', 'yearmonthdate(date):T', 'record_count:Q']
).interactive().properties(width=800, height=500)

chart.display()
chart.save('MJP-works-in-poetry-and-the-egoist-by-genre.html')
```

## Ex 6: Making an interactive scatterplot to explore the Seattle Library Circulation Data 2010-2015

For this example, we're going to use data from Research Exercise # 4 on the [Seattle Public Library Borrowing Records](https://github.com/sceckert/Data-and-Literary-Study-Spring2022/blob/main/_week4/research-exercise-4.ipynb)

### Let's start by filtering our dataframe to look at just works by a single author

```python
morrison_seattle_library_df = seattle_library_df[seattle_library_df['Author'].str.contains('Morrison, Toni', na=False)]
morrison_seattle_library_df 
```

### Plotting a scatterplot of checkouts of books by a single author in the Seattle Library data

Hypothesis: Checkouts of books roughly correlate with 

NOTE: Because our dataset is only checkouts of books that have been checked out *at least* 10 times in a month, we're not seeing all of the Toni Morrison data from 2010-2015. 

```python
alt.Chart(morrison_seattle_library_df,\
          title='How Many Checkouts of Toni Morrison Books at the Seattle Public Library?').mark_circle(size=40).\
encode(
    x=alt.X('yearmonth(CheckoutYear_CheckoutMonth):T', axis=alt.Axis(format='', title='Checkout Year')
    ),
    y=alt.Y('Checkouts:Q', axis=alt.Axis(title='Number of checkouts')),
    tooltip=['Title', 'PublicationYear', 'yearmonth(CheckoutYear_CheckoutMonth)', alt.Tooltip('Checkouts')]
).interactive().properties(width=800, height=500)
```

### Plotting a scatterplot of checkouts of ALL books in Seattle Library data 2010-2015

```python
# Altair normally sets 5,000 as the maximum number of rows in a dataset
# Because our Seattle Public Library Dataset is so massive (434873 rows!), we need to temporarily disable this cap
alt.data_transformers.disable_max_rows()

# Create our altair chart
alt.Chart(seattle_library_df, title='How Many Checkouts at the Seattle Public Library?').mark_circle(size=40).encode(
   x=alt.X('yearmonth(CheckoutYear_CheckoutMonth):T', axis=alt.Axis(format='', title='Checkout Year')
    ),
    y=alt.Y('Checkouts:Q', axis=alt.Axis(title='Number of checkouts')),
    tooltip=['Title', 'PublicationYear','CheckoutYear_CheckoutMonth', alt.Tooltip('Checkouts:Q')]
).interactive().properties(width=800, height=500)
```

## Bonus: Create a Gantt chart to visualize intervals over time in the Literature Humanities Texts 1937-2020 Data
We're going to revisit the [Literature Humanities](https://www.college.columbia.edu/core/1937.php) Syllabus [data](https://github.com/sceckert/Data-and-Literary-Study-Spring2022/blob/main/_datasets/Columbia-Literature-Humanities-Syllabi/columbia-university-literature-humanities-syllabus-texts-1937-2020-cleaned.csv) from [Research Exercise #5](https://github.com/sceckert/Data-and-Literary-Study-Spring2022/blob/main/_week5/research-exercise-5.md#2-columbia-universitys-literature-humanities-syllabus-1937-2020). If you don't remember where this data came from, take a peek back at the research exercise and links in the description. 

```python
lit_hum_df = pd.read_csv('../_datasets/Columbia-Literature-Humanities-Syllabi/columbia-university-literature-humanities-syllabus-texts-1937-2020-cleaned.csv', encoding='utf-8')
```

```python
lit_hum_df.head(5)
```

```python
# Because some of the texts, like "Confessions," have the same title,
# let's create a column of uniqke idefiters and call it "Author_and_Text"
lit_hum_df['Author_and_Text'] = lit_hum_df['Author'] + ', ' + lit_hum_df['Text']
```

### Use Altair's to a Gantt chart to visualize intervals over time
Here, we're again using the `.mark_bar()` operation that we used to create the stacked bar charts. But this time, notice that we're encoding TWO sets of x-axis data (the academic years START and END dates)


Also note that we're not using the ":T" encoding parameter, because our year data is not in YYY-MM-DD format. Instead, we're encoding it as type ":O" (ordinal numeric data).

```python
chart= alt.Chart(lit_hum_df,\
                 title="Literature Humanities Texts by Years Assigned, 1937-2020").mark_bar().\
encode(
    x=alt.X('AcademicYearsStart:O', axis=alt.Axis(title='Years Assigned')),
    x2='AcademicYearsEnd:O',
    y='Author_and_Text'
)
chart.save('LitHUm-Texts-by-Years-Assigned-1937-2020.html')
chart
```
