### Bonus: Data cleaning
Yikes! Our chart is a little hard to see. If you right click on the image and open it in a new tab (or zoom in your browser) you'll see that the culprit is the long title for Thomas Paine's eighteenth-century treatise, *Common Sense: Addressed to the Inhabitants of America, on the Following Interesting Subjects: I. IV. Of the Present Ability of America, With Some Miscellaneous Reflections*.

If we want to visualize our data in a slightly more legible format, we can do that without overwriting our data. Let's create a new column with cleaned titles called "Titles_cleaned" and use this new column to create our visualization.

First, let's create an empty list, called "Titles_cleaned." Then, let's create a `for` loop to loop over the items in the titles "Titles". Using an `if` statement, if a given title contain the string "Common Sense", instead of copying the title over, add the string "Common Sense".  Otherwise (an `else` statment), copy the title over to our list.

Like so:

```python
#create an empty list called "titles cleaned"
titles_cleaned = []

# Use a for loop to loop over the titles in our Titles column  and copy most of them over to our "titles_cleaned" list
for title in open_syllabus_project_df['Titles']:
    if 'Common Sense' in title:
        titles_cleaned.append('Common Sense')
    else:
        titles_cleaned.append(title)
```

```python
#Let's check our list
titles_cleaned
```

Now, we're going to make a new column in our dataframe with this list:

```python
open_syllabus_project_df['Titles_cleaned'] = titles_cleaned
open_syllabus_project_df
```

Now, let's try replotting!

```python
open_syllabus_project_df.plot.barh(x='Titles_cleaned', y="Coassignments", figsize=(20,20), title="Texts Coassigned with *The Souls of Black Folk* \in Open Syallbus Project Syllabi")
```

## Exercise 2: Parsing XML versions of *The Souls of Black Folk*

Together with your partner, load in the XML file of Du Bois's *The Souls of Black Folk*. Extract just the text of all the epigraphs, then, put that output into its own text file. 


```python
# Your code here
from bs4 import BeautifulSoup

tei_dubois = '../_datasets/texts/literature/du-bois-the-soul-of-black-folk.xml'

with open(tei_dubois, 'r', encoding='utf-8') as tei:
    soup = BeautifulSoup(tei, 'lxml')
```

```python
#Create a new blank list called "epigraphs"
epigraphs = []

# Use a for loop to iterate over all the <epigraph> tags in the BeautifulSoup object, 
#  using the .get_text() operation to strip the tags
# and use the .append() function to append each epigraph to the list we defined above ("epigraphs")
for entry in soup.find_all('epigraph'):
    epigraphs.append(entry.get_text())
```

```python
# Let's check to make sure our epigraphs were all added
print(epigraphs)
```

```python
# Let's write our list to a text file, using the ususal .open() function
# Because we have a list, rather than a string, we'll need to use the `.join()` command 
# and the newline character ("\n") to join each item in our epigraphs list, separated by a new line.
open('souls-of-black-folks-epigraphs.txt', mode='w', encoding='utf-8').write("\n".join(epigraphs))
```

## Exercise 4: Measuring the duration of titles on Literature Humanities syllabi

Together with your partner, brainstorm: how would figure out what title has ben taught on the Literature Humanities dataset the longest?

One way we might figure out what work has been taught the longest is to add up the duration of all years it has been taught. We can perform mathe on numerical columns in pandas (like add and subract columns). We can also set the output equal to a new column.We could create a new column whose values include the number of years a text was taught in a given year. 


Once we've done that, we'll have one more step to do: Because some texts were taught at intermitten intervals, we'll have to use `.groupby()` text (and the `.sum()` function) to get the total number of years taught.


Let's start by reading in our dataframe:

```python
literature_humanities_df = pd.read_csv('../_datasets/Columbia-Literature-Humanities-Syllabi/columbia-university-literature-humanities-syllabus-texts-1937-2020-cleaned.csv')
```

```python
literature_humanities_df
```

```python
#Let's create 
start_times = literature_humanities_df["AcademicYearsStart"]
end_times = literature_humanities_df["AcademicYearsEnd"]

# Let's create a new column in our dataframe whose values are the difference between the 
# Academic Year Start and End dates. 
literature_humanities_df['NumberOfYearsTaught'] = end_times - start_times 
```

```python
literature_humanities_df
```

```python
# Let's use .groupby() to group by texts and authorsand sum up the total number of years taught:
# Why text *and* author? Because some texts have the same name (See Augustine and Rousseau's "Confessions")
literature_humanities_df.groupby(['Text', 'Author'])['NumberOfYearsTaught'].sum().sort_values(ascending=False)
```

```python
# Let's see what our data plot!

literature_humanities_df.groupby(['Text', 'Author'])['NumberOfYearsTaught'].sum().sort_values(ascending=False)\
.plot.barh(x='Text', y='NumberOfYearsTaught', figsize=(20,25),\
           title="How Many Years Have Texts Been Taught on the \nColumbia Literature Humanites Syllabus, 1973-2020")
```
